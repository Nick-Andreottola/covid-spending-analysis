## Question 2

```python
years_2020 = data[data[:,0] == 2020]
print(years_2020)
```

```python
ACF_spending = years_2020[:,7]
AER_spending = years_2020[:,8]
GEN_AAP_spending = years_2020[:,9]
GRF_spending = years_2020[:,10]
HCS_spending = years_2020[:,11]
TWS_spending = years_2020[:,12]
Retail_and_Grocery_spending = years_2020[:,13]
Retail_spending = years_2020[:,14]

print("ACF_spending: ", np.sum(ACF_spending > 0))
print("AER_spending: ", np.sum(AER_spending > 0))
print("GEN_AAP_spending: ", np.sum(GEN_AAP_spending > 0))
print("GRF_spending: ", np.sum(GRF_spending > 0))
print("HCS_spending: ", np.sum(HCS_spending > 0))
print("TWS_spending: ", np.sum(TWS_spending > 0))
print("Retail_and_Grocery_spending: ", np.sum(Retail_and_Grocery_spending > 0))
print("Retail_spending: ", np.sum(Retail_spending > 0))
```

The category that spent the most days above 0 is grocery and food store which makes sense because people needed food during covid and wanted to stock up. The category that spent the fewest days above 0 is transportation and warehousing which makes sense because people weren't traveling during covid for fear of getting sick.

## Question 3

```python
years_2020_3 = years_2020[years_2020[:,1] == 3]
X = years_2020_3[:,2]
Y1 = years_2020_3[:,8]

Y2 = years_2020_3[:,10]


plt.plot(X,Y1, label = 'arts')
plt.plot(X,Y2, label = 'grocery')
plt.title('Arts, entertainment, and recreation (AER) and Grocery and food store spending (GRF)')
plt.xlabel('Day')
plt.ylabel('Spending value')
plt.legend()
plt.show()


```

My approach was to try and pick two things I figured would show polar opposite results graphically. For example, I knew the graph of GRF would have a significant peak because it would have been popular during covid. While things like arts and recreation would not have as high a spending value. I used plt to add a legend, labels, and a title to make it clear what the graph was showing.
